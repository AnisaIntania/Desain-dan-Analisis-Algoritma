**3 Studi Kasus untuk implementasi DFS dan BFS**

**1. AMIN - BFS**

```python
graph = {'Amin' : ['Wasim', 'Nick', 'Mike'],
         'Wasim' : ['Imran', 'Amin'],
         'Imran' : ['Wasim', 'Faras'],
         'Faras' : ['Imran'],
         'Mike' : ['Amin'],
         'Nick' : ['Amin']
         }

def bfs(graph, start):
    visited = []
    queue = []
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            neighbours = graph[node]
            for neighbour in neighbours:
                queue.append(neighbour)
    return visited
```

```python
# O Amin --------------------------
# |               |               |
# O Wasim       O Nick          O Mike
# |
# O Imran
# |
# O Faras

bfs (graph, 'Amin')
```

```python
bfs (graph, 'Wasim')
```

```python
bfs (graph, 'Faras')
```

**1. AMIN - DFS**

```python
graph = {
    'Amin' : ['Wasim', 'Nick', 'Mike'],
    'Wasim' : ['Imran', 'Amin'],
    'Imran' : ['Wasim', 'Faras'],
    'Faras' : ['Imran'],
    'Mike' : ['Amin'],
    'Nick' : ['Amin']
}

def dfs(graph, start, visited=None):
    if visited is None:
        visited = []

    visited.append(start)

    for neighbour in graph[start]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)

    return visited
```

```python
# O Amin --------------------------
# |               |               |
# O Wasim       O Nick          O Mike
# |
# O Imran
# |
# O Faras

dfs (graph, 'Amin')
```

```python
dfs (graph, 'Wasim')
```

```python
dfs (graph, 'Faras')
```

**2. REKTOR - BFS**

```python
from collections import deque

graph = {
    "Rektor": ["Warek 1", "Warek 2"],
    "Warek 1": ["Kaprodi 1"],
    "Warek 2": ["Kaprodi 2", "Kaprodi 3"],
    "Kaprodi 1": ["Dosen A", "Dosen B"],
    "Kaprodi 2": ["Dosen D", "Dosen E"],
    "Kaprodi 3": ["Dosen F", "Dosen G"],
    "Dosen A": [],
    "Dosen B": ["Dosen C"],
    "Dosen C": [],
    "Dosen D": [],
    "Dosen E": [],
    "Dosen F": [],
    "Dosen G": []
}

def bfs(start, graph):
    visited = []
    queue = deque([start])
    while queue:
        node = queue.popleft()
        visited.append(node)
        for n in graph[node]:
            queue.append(n)
    return visited

hasil_bfs = bfs("Rektor", graph)

print("Hasil BFS:")
for item in hasil_bfs:
    print(item)
```

**2. REKTOR - DFS**

```python
graph = {
    "Rektor": ["Warek 1", "Warek 2"],
    "Warek 1": ["Kaprodi 1"],
    "Warek 2": ["Kaprodi 2", "Kaprodi 3"],
    "Kaprodi 1": ["Dosen A", "Dosen B"],
    "Kaprodi 2": ["Dosen D", "Dosen E"],
    "Kaprodi 3": ["Dosen F", "Dosen G"],
    "Dosen A": [],
    "Dosen B": ["Dosen C"],
    "Dosen C": [],
    "Dosen D": [],
    "Dosen E": [],
    "Dosen F": [],
    "Dosen G": []
}

def dfs(start, graph, visited=None):
    if visited is None:
        visited = []
    visited.append(start)
    for n in graph[start]:
        dfs(n, graph, visited)
    return visited

hasil_dfs = dfs("Rektor", graph)

print("Hasil DFS:")
for item in hasil_dfs:
    print(item)
```

**3. TITIK 0 - BFS**

```python
graph0 = {
    0: [1, 2],
    1: [3, 4],
    2: [5],
    3: [],
    4: [6],
    5: [],
    6: []
}

def bfs(start, graph):
    visited = []
    queue = deque([start])
    while queue:
        node = queue.popleft()
        visited.append(node)
        for n in graph[node]:
            queue.append(n)
    return visited

hasil_bfs = bfs(0, graph0)

print("Hasil BFS:")
print(" → ".join(str(i) for i in hasil_bfs))
```

**3. TITIK 0 - DFS**

```python
graph0 = {
    0: [1, 2],
    1: [3, 4],
    2: [5],
    3: [],
    4: [6],
    5: [],
    6: []
}

def dfs(start, graph, visited=None):
    if visited is None:
        visited = []
    visited.append(start)
    for n in graph[start]:
        dfs(n, graph, visited)
    return visited

hasil_dfs = dfs(0, graph0)

print("Hasil DFS:")
print(" → ".join(str(i) for i in hasil_dfs))
```
