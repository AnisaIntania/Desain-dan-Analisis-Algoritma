**SOAL 1 — Hitung Jumlah Angka Genap**

```python
def hitung_genap(arr):
    count = 0
    for num in arr:
        if num % 2 == 0:
            count += 1
    return count

data = input("Masukkan angka (pisahkan dengan spasi): ")
arr = list(map(int, data.split()))
print("Jumlah angka genap:", hitung_genap(arr))
```

**Algoritma:** Algoritma dimulai dengan
membuat counter bernilai 0, lalu menelusuri seluruh elemen array satu per satu. Setiap angka dicek apakah genap menggunakan operasi modulus 2. Jika genap, counter ditambah. Setelah loop selesai, counter dikembalikan sebagai hasil

**Edge Cases:** jika array yang diberikan kosong maka fungsi akan menghasilkan nilai 0 karena tidak ada angka yang dapat diperiksa. Jika array berisi hanya bilangan ganjil, hasilnya juga tetap 0 karena tidak ditemukan angka genap. Selain itu, perlu diperhatikan bahwa bilangan negatif tetap dapat dihitung sebagai genap selama habis dibagi dua, misalnya -4 tetap dianggap sebagai angka genap.

**SOAL 2 — Cari Nilai Terbesar**

```python
def cari_maksimum(arr):
    if not arr:
        return "Array kosong"
    maks = arr[0]
    for num in arr:
        if num > maks:
            maks = num
    return maks

data = input("Masukkan angka (pisahkan dengan spasi): ")
arr = list(map(int, data.split()))
print("Nilai terbesar:", cari_maksimum(arr))
```

1. Langkahnya dimulai dengan menganggap elemen pertama sebagai nilai terbesar sementara. Kemudian setiap elemen dibandingkan satu per satu. Jika ada angka yang lebih besar, nilai maksimum diperbarui.

2. Jika array kosong, fungsi dapat mengembalikan None atau pesan khusus.

**SOAL 3 — Balik String**

```python
def balik_string(s):
    return s[::-1]

text = input("Masukkan kata: ")
print("Hasil dibalik:", balik_string(text))
```

String dibalik dengan mengambil karakter dari belakang ke depan, bisa menggunakan teknik slicing atau menggabungkan karakter satu per satu dalam urutan terbalik. Hasil akhir berupa string baru yang urutannya terbalik dari input awal.

**SOAL 4 — Jumlahkan Semua Elemen**

```python
def jumlah_elemen(arr):
    total = 0
    for num in arr:
        total += num
    return total

data = input("Masukkan angka (pisahkan dengan spasi): ")
arr = list(map(int, data.split()))
print("Jumlah total elemen:", jumlah_elemen(arr))
```

Fungsi ini bekerja dengan membuat variabel total bernilai 0, lalu menjumlahkan setiap angka di dalam array secara bertahap. Setelah seluruh angka ditambahkan, nilai total dikembalikan sebagai hasil akhir.

**SOAL 5 — Cek Palindrome (String)**

```python
def cek_palindrome(text):
    return text == text[::-1]

text = input("Masukkan kata: ")
if cek_palindrome(text):
    print("Palindrome")
else:
    print("Bukan palindrome")
```

String dianggap palindrome jika dibaca sama dari depan maupun belakang. Caranya adalah dengan membandingkan string asli dengan versi terbaliknya. Jika keduanya sama, maka string tersebut palindrome.

**SOAL 6 — Hitung Jumlah Huruf Vokal**

```python
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    count = 0
    for char in s:
        if char in vokal:
            count += 1
    return count

text = input("Masukkan kata: ")
print("Jumlah huruf vokal:", hitung_vokal(text))
```

Setiap karakter dalam string dicek apakah termasuk huruf vokal (a, i, u, e, o) dengan membandingkannya ke daftar vokal. Jika cocok, counter ditambah. Di akhir, fungsi mengembalikan jumlah vokal yang ditemukan.

**SOAL 7 — Temukan Elemen yang Dicari**

```python
def cari_target(arr, target):
    for num in arr:
        if num == target:
            return True
    return False

data = input("Masukkan angka (pisahkan dengan spasi): ")
arr = list(map(int, data.split()))
target = int(input("Masukkan angka yang dicari: "))

if cari_target(arr, target):
    print("Ada")
else:
    print("Tidak ada")
```

Pencarian dilakukan dengan memeriksa setiap elemen array satu per satu dan membandingkannya dengan target. Jika ditemukan kecocokan, fungsi langsung mengembalikan True. Jika tidak ditemukan hingga akhir, hasilnya False.
