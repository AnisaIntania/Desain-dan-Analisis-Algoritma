```python
for y in range(0, 3):
    x = 7 - 3*y
    print(f"x = {x}, y = {y}")
```

```python
from scipy.optimize import linprog

# Maksimalkan Z = 5x + 3y
# linprog hanya bisa MINIMIZE, jadi kita ubah jadi: minimize -Z
c = [-5, -3]

# Batasan: 4x + 12y <= 28
A = [[4, 12]]
b = [28]

# Batas bawah (x>=0, y>=0)
x_bounds = (0, None)
y_bounds = (0, None)

result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])

print("Nilai optimal:")
print(f"x = {result.x[0]:.2f}")
print(f"y = {result.x[1]:.2f}")
print(f"Z maks = {-result.fun:.2f}")
```

```python
for x in range(0, 11):  # coba nilai x dari 0 sampai 10
    y = 21 - 2 * x
    print(f"x = {x}, y = {y}")
```

```python
from scipy.optimize import linprog

c = [-6, -4]  # karena mau maksimalkan
A = [[2, 1]]
b = [21]
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A, b_ub=b, bounds=bounds)
print(f"x = {res.x[0]:.2f}")
print(f"y = {res.x[1]:.2f}")
print(f"Z maks = {-res.fun:.2f}")
```

```python
from pulp import LpProblem, LpVariable, LpMinimize

# Membuat model Linear Programming
model = LpProblem("Masalah_Lisa_dan_Muri", LpMinimize)

# Variabel keputusan: jam kerja Lisa (L) dan Muri (M)
L = LpVariable('L', lowBound=0)
M = LpVariable('M', lowBound=0)

# Fungsi tujuan (ga perlu optimasi, jadi kita buat 0)
model += 0, "Fungsi_Tujuan"

# Kendala:
model += L + M == 16, "Total_Jam"
model += 3*L + 4*M == 55, "Total_Tas"

# Jalankan solver bawaan PuLP (CBC)
model.solve()

# Tampilkan hasil
print("=== HASIL LINEAR PROGRAMMING (PuLP) ===")
print(f"Lisa (L) = {L.value():.0f} jam")
print(f"Muri (M) = {M.value():.0f} jam")
```
